Copy vs View

```python
a = np.array([1,2,3])
b = a.copy()
b[0] = 100
print(a)
print(b)
```

```python
a = np.array([1,2,3])
b= a.view()
b[1]= 100
print(a)
```

```python
m = np.array([5,6,7])
n = m.copy()
v = m.view()
print(m)
print(n)
print(v)
```

Sorting

```python
arr = np.array([5,2,6,8,35,1])
b = np.sort(arr)
print(b)

```

```python
#find 2nd max num ijn array
arr = np.array([5,2,6,8,35,1])
b = np.sort(arr)
c = np.max(b[-2])
print(b)
print(c)
```

Boolean filtering

```python
arr = np.array([10,20,30,40,50])

print(arr>25)
print(arr[arr>25])
```

```python
#problem
arr = np.array([12,45,7,89,23])
print(arr)
print(np.max(arr)) #max
print(np.min(arr)) #min
print(np.average(arr)) #avg
```

```python
#find max element without using np.max()
arr = np.array([8, 3, 15, 2, 20, 10])
a=np.sort(arr)
b = a[-1]
print(a)
print(b)

m = np.sort(arr)[-1]
print(m)
```

```python
import numpy as np

arr = np.array([12, 45, 7, 89, 23])
max_num = arr[0]
for i in arr:
    if i > max_num:
        max_num = i

print(max_num)
```
